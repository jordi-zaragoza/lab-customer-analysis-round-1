```python
import pandas as pd 
import numpy as np
```

```python
file1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
```

```python
cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols

cols = []
for column in file2.columns:
    cols.append(column.lower())
file2.columns = cols

cols = []
for column in file3.columns:
    cols.append(column.lower())
file3.columns = cols

file3 = file3.rename(columns={'state':'st'})
```

```python
file1 = file1[['customer', 'st', 'gender','education','customer lifetime value','income', 'monthly premium auto', 'number of open complaints', 'total claim amount', 'policy type', 'vehicle class']]
display(file1.head(3))
file1.shape
```

```python
file2 = file2[['customer', 'st', 'gender','education','customer lifetime value','income', 'monthly premium auto', 'number of open complaints', 'total claim amount', 'policy type', 'vehicle class']]
display(file2.head(3))
file2.shape
```

```python
file3 = file3[['customer', 'st', 'gender','education','customer lifetime value','income', 'monthly premium auto', 'number of open complaints', 'total claim amount', 'policy type', 'vehicle class']]
display(file3.head(3))
file3.shape
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data = data.reset_index(drop = True)
data
```

```python
data.select_dtypes(include=['integer','float'])
```

```python
data.select_dtypes(exclude=['integer','float'])
```

```python
data = data.drop(['number of open complaints','education'], axis=1)
```

```python
data['customer lifetime value'] =  pd.to_numeric(data['customer lifetime value'], errors='coerce')*100
```
